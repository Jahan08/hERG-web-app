```
import pandas as pd

import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import PandasTools

print(rdkit.rdBase.rdkitVersion)
```

```
world_drugs =pd.read_csv("/content/bioactivity_DprE1_Mtb_IC50_updated.csv")
```

```
PandasTools.AddMoleculeColumnToFrame(world_drugs,'canonical_SMILES', 'Molecule')
PandasTools.WriteSDF(world_drugs, 'DprE1_Mtb_IC50_new.sdf', molColName='Molecule', properties=list(world_drugs.columns))
```
